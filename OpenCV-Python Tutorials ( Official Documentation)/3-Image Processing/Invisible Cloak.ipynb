```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

ret, background = cap.read()
background = np.flip(background, axis = 1)
cv2.imshow("Background",background)

while True:

    _, frame = cap.read()
    frame = np.flip(frame, axis = 1)

    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    cv2.imshow("hsv",hsv)
 
    #red
    lower1 = np.array(([165, 170, 130]))
    upper1 = np.array(([255, 255, 255]))
    
    #blue
    low_blue = np.array([70, 170, 130])
    high_blue = np.array([160, 255, 255])
    
    mask_blue = cv2.inRange(hsv,low_blue,high_blue)
    mask_red = cv2.inRange(hsv,lower1,upper1)

    
    kernel = np.ones((5,5),np.uint8)
    closing = cv2.morphologyEx(mask_red, cv2.MORPH_CLOSE,kernel)
    n_closing = np.bitwise_not(closing)


    res_c = cv2.bitwise_and(background, background, mask = closing)
    res = cv2.bitwise_and(frame,frame,mask = n_closing)

    final = cv2.addWeighted(res,1,res_c,1,0)
    blur = cv2.medianBlur(final,7)
    cv2.imshow("final",blur)

    k = cv2.waitKey(5) & 0xFF
    if k == 27:
        break
        
cap.release()
cv2.imshow("background",background)
cv2.destroyAllWindows()
```
